#Part 2: Now, rather than passing it all into a function, we can do all that with a decorator. 

```
#Example: 
#1. Conventional Approach

def adder(x,y):
  return x+y

interact(adder,x=5,y=5)
```

```
#Doing it with decorators
@interact(x=5,y=5)

def adder(x,y):
  return x+y
```

###The **var =fixed( value)** function fixes the var value and hence we cannot make any changes/toggle/play with that value anymore

```
#Doing it with decorators and fixing one value
@interact(x=5,y=fixed(5))

def adder(x,y):
  return x+y

#you can see below that this is just adding 5 to the value of x
```

#Example of what a GUI WIDGET can do

```

%matplotlib inline

from ipywidgets import interact, interactive
from IPython.display import clear_output, display, HTML

import numpy as np
from scipy import integrate

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import cnames
from matplotlib import animation
```

```
def solve_lorenz(N=10, angle=0.0, max_time=4.0, sigma=10.0, beta=8./3, rho=28.0):

    fig = plt.figure();
    ax = fig.add_axes([0, 0, 1, 1], projection='3d');
    ax.axis('off')

    # prepare the axes limits
    ax.set_xlim((-25, 25))
    ax.set_ylim((-35, 35))
    ax.set_zlim((5, 55))
    
    def lorenz_deriv(x_y_z, t0, sigma=sigma, beta=beta, rho=rho):
        """Compute the time-derivative of a Lorenz system."""
        x, y, z = x_y_z
        return [sigma * (y - x), x * (rho - z) - y, x * y - beta * z]

    # Choose random starting points, uniformly distributed from -15 to 15
    np.random.seed(1)
    x0 = -15 + 30 * np.random.random((N, 3))

    # Solve for the trajectories
    t = np.linspace(0, max_time, int(250*max_time))
    x_t = np.asarray([integrate.odeint(lorenz_deriv, x0i, t)
                      for x0i in x0])
    
    # choose a different color for each trajectory
    colors = plt.cm.jet(np.linspace(0, 1, N));

    for i in range(N):
        x, y, z = x_t[i,:,:].T
        lines = ax.plot(x, y, z, '-', c=colors[i])
        _ = plt.setp(lines, linewidth=2);

    ax.view_init(30, angle)
    _ = plt.show();

    return t, x_t
```

```
w4 = interactive(solve_lorenz, angle=(0.,360.), N=(0,50), sigma=(0.0,50.0), rho=(0.0,50.0))
display(w4);
```

```
#SOlving an oil and gas problem with interactive widgets. 

def flow_profiler(mu):
  r = np.linspace(2000,0.1)
  p = 3000 - (141.2*200*mu*1.1*np.log(2000/r)/(350*30))
  plt.plot(r,p)
  plt.xlim(0.1,2000)
  plt.ylim(500,3000)
  plt.xlabel('Radius')
  plt.ylabel('Pressure')
  plt.title('Flow Profile variation with viscosity')
  return p,r
```

```
w_res = interactive(flow_profiler,mu=[100,1000])
```
